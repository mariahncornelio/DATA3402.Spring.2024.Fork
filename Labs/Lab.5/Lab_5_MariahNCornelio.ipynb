1. Create a `matrix` class with the following properties:
    * It can be initialized in 2 ways:
        1. with arguments `n` and `m`, the size of the matrix. A newly instanciated matrix will contain all zeros.
        2. with a list of lists of values. Note that since we are using lists of lists to implement matrices, it is possible that not all rows have the same number of columns. Test explicitly that the matrix is properly specified.
    * Matrix instances `M` can be indexed with `M[i][j]` and `M[i,j]`.
    * Matrix assignment works in 2 ways:
        1. If `M_1` and `M_2` are `matrix` instances `M_1=M_2` sets the values of `M_1` to those of `M_2`, if they are the same size. Error otherwise.
        2. In example above `M_2` can be a list of lists of correct size.


```python
class matrix:
    def __init__(self, n, m=None, list_matrix=None):
        if m==0 or m is None:
            m=n # This creates a n x n matrix if no value for m is given
            
        if list_matrix is None: # If no list of lists is provided
            self.rows=[[0]*m for i in range(n)]
        else:
            # This part checks if what is provided is a valid list and if it is the same length
            if all(isinstance(row, list) for row in list_matrix) and all(len(row) == len(list_matrix[0]) for row in list_matrix[1:]):
                self.rows=list_matrix
            else:
                raise ValueError("Invalid matrix. All rows must have the same number of columns OR data given needs to be a list.")

    def __getitem__(self, position): # Indexing matrices with M[i][j] or M[i,j]
        # Used to be def(index_matrix(self, position): but it would output a TypeError error 
        if isinstance(position, tuple) and len(position)==2:
            i,j=position
            return self.rows[i][j]
        elif isinstance(position, int):
            return self.rows[position]
        else:
            raise IndexError("Invalid index. Use M[i, j] or M[i][j].")

    def __repr__(self): # This __repr__ function helps turn matrix into string
        return repr(self.rows)

    def assign_matrix(self, matrix2):
        if len(self.rows)==len(matrix2.rows) and all(len(row)==len(matrix2.rows[0]) for row in self.rows):
            self.rows=[row[:] for row in matrix2.rows]
        else:
            raise ValueError("Matrix assignment error. Matrices must be the same size.")
```

3. Write functions that create special matrices (note these are standalone functions, not member functions of your `matrix` class):
    * `constant(n,m,c)`: returns a `n` by `m` matrix filled with floats of value `c`.
    * `zeros(n,m)` and `ones(n,m)`: return `n` by `m` matrices filled with floats of value `0` and `1`, respectively.
    * `eye(n)`: returns the n by n identity matrix.

```python
def constant(n, m, c):
     return matrix(n, m, list_matrix=[[c]*m for i in range(n)])

def zeros(n, m):
     return constant(n, m, 0)

def ones(n, m):
     return constant(n, m, 1)

# An identity matrix essentially looks like this (I am using a 3x3 matrix as an example):
# 1 0 0 
# 0 1 0 
# 0 0 1

def eye(n):
    identity=zeros(n, n)
    for i in range(n):
        identity[i][i]=1 # This places 1's in the diagonal
    return identity
```

```python
class matrix:
    # This part of the code is the same as the previous question
    def __init__(self, n, m=None, list_matrix=None):
        if m==0 or m is None:
            m=n
        if list_matrix is None:
            self.rows=[[0]*m for i in range(n)]
        else:
            if all(isinstance(row, list) for row in list_matrix) and all(len(row) == len(list_matrix[0]) for row in list_matrix[1:]):
                self.rows=list_matrix
            else:
                raise ValueError("Invalid matrix. All rows must have the same number of columns OR data given needs to be a list.")
    def __repr__(self):
        return repr(self.rows)
    def assign_matrix(self, matrix2):
        if len(self.rows)==len(matrix2.rows) and all(len(row)==len(matrix2.rows[0]) for row in self.rows):
            self.rows=[row[:] for row in matrix2.rows]
        else:
            raise ValueError("Matrix assignment error. Matrices must be the same size.")
    def shape(self):
        return len(self.rows), len(self.rows[0])
    def transpose(self):
        return matrix(len(self.rows[0]), list_matrix=[[self.rows[j][i] for j in range(len(self.rows))] for i in range(len(self.rows[0]))])
    def row(self, n):
        return matrix(1, len(self.rows[0]), list_matrix=[self.rows[n]])  
    def column(self, n):
        return matrix(len(self.rows), 1, list_matrix=[[self.rows[i][n]] for i in range(len(self.rows))])
    def to_list(self):
        return self.rows
    def block(self, n_0, n_1, m_0, m_1):
        return matrix(n_1-n_0+1, m_1-m_0+1, list_matrix=[row[m_0:m_1+1] for row in self.rows[n_0:n_1+1]])
    def __getitem__(self, position):
        if isinstance(position, tuple) and len(position) == 2:
            i,j=position
            return self.rows[i][j]
        elif isinstance(position, int):
            return self.rows[position]
        elif isinstance(position, slice):
            return [row[:] for row in self.rows[position]]
        else:
            raise IndexError("Invalid index. Use M[i, j], M[i][j], or M[index1:index2].")

    # My solution to question 4
    def scalarmul(self, c):
        return matrix(len(self.rows), len(self.rows[0]), list_matrix=[[c*self.rows[i][j] for j in range(len(self.rows[0]))] for i in range(len(self.rows))])

    def add(self, N):
         # To add and subtract 2 matrices, you need to make sure that they are the same dimensions
        if len(self.rows)==len(N.rows) and len(self.rows[0])==len(N.rows[0]):
            return matrix(len(self.rows), len(self.rows[0]), list_matrix=[[self.rows[i][j]+N.rows[i][j] for j in range(len(self.rows[0]))] for i in range(len(self.rows))])
        else:
            raise ValueError("ERROR. Matrices must be the same dimensions.")

    def sub(self, N):
        # To add and subtract 2 matrices, you need to make sure that they are the same dimensions
        if len(self.rows)==len(N.rows) and len(self.rows[0])==len(N.rows[0]):
            return matrix(len(self.rows), len(self.rows[0]), list_matrix=[[self.rows[i][j]-N.rows[i][j] for j in range(len(self.rows[0]))] for i in range(len(self.rows))])
        else:
            raise ValueError("ERROR. Matrices must be the same size.")

    def mat_mult(self, N):
        # To multiply 2 matrices/get the dot product, you need to make sure the inside number of
        # n x m of matrix A and n x m of matrix B are the same. So it will work if the matrix 
        # dimensions are 2 x 3 and 3 x 5 but not if it is 2 x 3 and 4 x 6 
        if len(self.rows[0])==len(N.rows): # Column number of matrix A need to match row number of matrix B
            dot_product=matrix(len(self.rows), len(N.rows[0]))
            for i in range(len(self.rows)):
                for j in range(len(N.rows[0])):
                    dot_product[i][j]=sum(self.rows[i][k]*N.rows[k][j] for k in range(len(self.rows[0])))
            return dot_product
        else:
            raise ValueError("ERROR. The number of columns in M must be the same as the number of rows in N.")

    def element_mult(self, N):
        if len(self.rows)==len(N.rows) and len(self.rows[0])==len(N.rows[0]):
            return matrix(len(self.rows), len(self.rows[0]), list_matrix=[[self.rows[i][j]*N.rows[i][j] for j in range(len(self.rows[0]))] for i in range(len(self.rows))])
        else:
            raise ValueError("ERROR. Matrices must be the same size.")

    def equals(self, N):
        if self.rows==N.rows:
            return True
        else:
            return False
```
